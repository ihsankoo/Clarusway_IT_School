Hello my dear students.  Here I am in Colab, starting fresh with a new Notebook.  What would I like to refresh my memory about?  Lets look at pandas Series type object.  

What I discover for myself in this worksheet is that I can feed the pandas DataFrame with a list of Series, in which case their names become rows.  

I might also name the Series inside a dictionary, using name : Series pairs.  In that case, the Series go in as columns, not as rows.

```python
import pandas as pd
from pandas import Series # also allowed
import numpy as np
```

```python
s1 = pd.Series(1) # is just a scalar OK?
```

```python
s1
```

```python
s1.name = "Space OK?"
```

```python
s1
```

```python
s2 = Series(2, index = ["a"], name = "Surprise me!")
```

```python
s2
```

Now lets see what effect Series names have on pandas DataFrame column names.

```python
df = pd.DataFrame([s1, s2]) # sequences side by side with no new information
```

```python
df
```

```python
s2
```

```python
s3 =s2.reset_index([0], drop=True)
```

```python
s4 = pd.Series(np.arange(4), name = "label4")
```

```python
s5 = pd.Series(np.arange(4,8), name = "label5")
```

```python
s5
```
