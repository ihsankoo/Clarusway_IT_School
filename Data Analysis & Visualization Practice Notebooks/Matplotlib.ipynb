Some matplotlib basics.

```python
import matplotlib as mpl
```

```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
domain = np.arange(-5, 5.1, 0.5)
```

```python
domain.shape
```

```python
len(domain)
```

```python
domain[-1]
```

```python
domain[-2]
```

```python
domain[-1] - domain[-2]
```

```python
y_values = domain * domain
```

```python
# plt.plot?
```

```python
plt.plot(domain, y_values);
```

```python
plt.plot(domain, y_values, 'gH--', 
         markersize=10, 
         markeredgecolor="m",
         markerfacecolor="m",
         label='line 1', linewidth=2);
```

[Setting axis range on Stackabuse](https://stackabuse.com/how-to-set-axis-range-xlim-ylim-in-matplotlib/)

[xticks](https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.xticks.html)

```python
fig, ax = plt.subplots()
plt.title("Parabola")
plt.xlabel("Domain")
plt.xlim(-6,6)
plt.xticks(np.arange(-6, 7))
plt.ylabel("Range")
plt.plot(domain, y_values, 'rp--', label='line 1', linewidth=2)
plt.legend()
plt.savefig("parabola.png", format="png");
```
